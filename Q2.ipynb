```python
import numpy as np
```

```python
from sklearn.datasets import fetch_20newsgroups

newsgroups = fetch_20newsgroups(subset="all", categories=["comp.sys.mac.hardware", "sci.space"])
x = np.array(newsgroups.data)
y = np.array(newsgroups.target)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words="english", lowercase=True, max_features=1000, binary=True)
vectorizer.fit(x)
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=1234)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.metrics import accuracy_score
```

```python
i = 1
for train_index, test_index in kf.split(x):
    clf = MultinomialNB(alpha=1.0)
    
    clf.fit(vectorizer.transform(x[train_index]), y[train_index])
    
    prediction = clf.predict(vectorizer.transform(x[test_index]))
    
    accuracy = accuracy_score(y[test_index], prediction)
    
    print("Accuracy in fold {}: {:.2%}.".format(i, accuracy))
    
    i += 1
```
